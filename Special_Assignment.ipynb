```python
sns.countplot(x=df["gender"])
plt.title("Gender Distribution")
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
```

```python
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

```python
sns.scatterplot(x="screen_time_before_sleep",y="stress_level",hue="gender",data=df)
plt.show()
```

```python
sns.boxplot(x="gender",y="stress_level",data=df)
plt.show()
```

```python
co=df.corr(numeric_only=True)
sns.heatmap(co,annot=True)
```
